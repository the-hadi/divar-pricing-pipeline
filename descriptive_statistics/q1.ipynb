## Data Loading

## Exploratory Data Analysis

### Listing Volume by Category

Distribution of existing advertisements across level-two (`cat2_slug`) and level-three (`cat3_slug`) categories.

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 14))

sns.countplot(
    data=df_raw,
    x="cat2_slug",
    ax=axes[0],
    order=df_raw["cat2_slug"].value_counts().index,
    palette="viridis",
)
axes[0].set_title("Category 2", fontsize=22, pad=15)
axes[0].set_ylabel("Counts", fontsize=12)
axes[0].set_xticklabels(axes[0].get_xticklabels(), rotation=30, ha="right")

sns.countplot(
    data=df_raw,
    x="cat3_slug",
    ax=axes[1],
    order=df_raw["cat3_slug"].value_counts().index,
    palette="magma",
)
axes[1].set_title("Category 3", fontsize=22, pad=15)
axes[1].set_ylabel("Counts", fontsize=12)
axes[1].set_xticklabels(axes[1].get_xticklabels(), rotation=45, ha="right")

plt.tight_layout()
plt.show()
```

### Key Insights from Category Distribution Analysis

* **Dominance of Residential Market:** In the **Category 2** breakdown, residential listings (`residential-sell` and `residential-rent`) overwhelmingly dominate the dataset, accounting for the vast majority of the total volume. `residential-sell` alone surpasses 500,000 listings, indicating a market highly geared toward property ownership advertisements.
* **Granular Breakdown (Category 3):** The **Category 3** plot reveals that apartments are the primary driver of this volume. `apartment-sell` is the single largest sub-category (over 300,000 counts), followed closely by `apartment-rent` (around 200,000 counts).
* **Traditional vs. Modern Residential:** Within the residential sector, older plots (`plot-old`) and villas/houses (`house-villa-sell` and `house-villa-rent`) maintain a substantial presence, though they remain significantly fewer than apartment listings.
* **Commercial and Niche Scarcity:** Commercial activities (`commercial-rent`, `commercial-sell`, `shop-rent`, `office-rent`) represent a much smaller fraction of the dataset. Specialized or emerging categories like `partnership`, `workspace`, and `industry-agriculture-business-sell` lie at the very long tail of the distribution, showing minimal volume in comparison to standard residential listings.
